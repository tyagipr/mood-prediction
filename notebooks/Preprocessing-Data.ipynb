```python
import math

#TODO  trying to clean up the missing values
mood_value_list = data['value'].tolist()
missing_value = [value for value in mood_value_list if value is None or math.isnan(value)]

#drop all the nan values as it was very less in numbers (202)
df_missing = pd.DataFrame(missing_value)
df_missing.dropna(inplace=True) 
print(len(df_missing))
print(len(data))
```

```python
rows_for_id = df[(df['id'] == 'AS14.01') & (df['variable'] == 'mood')]


# Print the filtered rows
print(rows_for_id.head())
```

```python
# fetching all rows with variable mood
all_rows = df[df['variable'] == 'mood']

# all the userid list
ids_list = all_rows['id'].tolist()
ids = set()

# Iterate through each dictionary in the list
for item in ids_list:
    # Add the ID to the set
    ids.add(item)

# Convert the set to a list
print(sorted(ids))
```

```python
# trying to write data in the new dataset
import csv

file = "/Users/pragyatyagi/PycharmProjects/mood-prediction/data/raw/mood-dataset-new.csv"
df['time'] = pd.to_datetime(df['time'])

    
# Open the CSV file in write mode
with open(file, 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['userid', 'date', 'mood'])

    for id_val in ids:
        filtered_df = df[(df['id'] == id_val) & (df['variable'] == 'mood')]

        average_df = round(df.groupby(filtered_df['time'].dt.date)['value'].mean())
        time_index = average_df.index

        date = time_index.astype(str).tolist()
        mood = average_df.tolist()
        userid = [id_val] * len(date)

        data = zip(userid, date, mood)

        # Write each tuple (date, value) to the CSV file
        csvwriter.writerows(data)





```
